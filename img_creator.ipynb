```python
from PIL import Image

def flip_image(image_path, saved_location):
    """
    Flip the image
    @param image_path: The path to the image to edit
    @param saved_location: Path to save the cropped image
    """
    image_obj = Image.open('/imagepath/imagename.jpg')
    rotated_image = image_obj.transpose(Image.FLIP_LEFT_RIGHT)
    rotated_image.save('/savedimages/f_imagename.jpg')
    rotated_image.show()
if __name__ == '__main__':
    image = 'imagename.jpg'
    flip_image(image, 'f_imagename.jpg')


def rotate1(image_path, degrees_to_rotate, saved_location):
    """
    Rotate the given photo with 90 degrees and save it
    @param image_path: The path to the image to edit
    @param degrees_to_rotate: The number of degrees to rotate the image
    @param saved_location: Path to save the cropped image
    """
    image_obj = Image.open('/imagepath/imagename.jpg')
    rotated_image = image_obj.rotate(90)
    rotated_image.save('/savedimages/r90_imagename.jpg')
    rotated_image.show()
if __name__ == '__main__':
    image = 'imagename.jpg'
    rotate1(image, 90, 'r90_imagename.jpg')


def rotate2(image_path, degrees_to_rotate, saved_location):
    """
    Rotate the given photo with 180 degrees and save it
    @param image_path: The path to the image to edit
    @param degrees_to_rotate: The number of degrees to rotate the image
    @param saved_location: Path to save the cropped image
    """
    image_obj = Image.open('/imagepath/imagename.jpg')
    rotated_image = image_obj.rotate(180)
    rotated_image.save('/savedimages/r180_imagename.jpg')
    rotated_image.show()
if __name__ == '__main__':
    image = 'imagename.jpg'
    rotate2(image, 180, 'r180_imagename.jpg')

def rotate3(image_path, degrees_to_rotate, saved_location):
    """
    Rotate the given photo with 270 degrees and save it
    @param image_path: The path to the image to edit
    @param degrees_to_rotate: The number of degrees to rotate the image
    @param saved_location: Path to save the cropped image
    """
    image_obj = Image.open('/imagepath/imagename.jpg')
    rotated_image = image_obj.rotate(270)
    rotated_image.save('/savedimages/r270_imagename.jpg')
    rotated_image.show()
if __name__ == '__main__':
    image = 'imagename.jpg'
    rotate3(image, 270, 'r270_imagename.jpg')
```
